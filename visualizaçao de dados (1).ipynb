um histograma usando o plotly.express, uma API para criação de gráficos no Python que contém diversos datasets de testes, identificados por data.xxx().

```python
 import plotly.express as px
 df = px.data.tips()
 fig = px.histogram(df, x='total_bill')
 fig.show()
```

Também podemos querer saber a correlação entre dados numéricos e, para isso, podemos utilizar o gráfico de dispersão, scatterplot, como no exemplo a seguir:

```python
import plotly.express as px
 df = px.data.iris()
 fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', symbol='species')
 fig.show()
```

Para dados categóricos, queremos mostrar as categorias e suas proporções, ou colorir outros gráficos como fizemos no scatterplot anteriormente. Por exemplo:

```python
import plotly.express as px
 long_df = px.data.medals_long()
 fig = px.bar(long_df, x='nation', y='count', color='medal', title='Long-Form Input')
 fig.show()
```

para entendermos proporções das categorias é o gráfico de pizza ou torta (pie).

```python
import plotly.express as px
 df = px.data.tips()
 fig = px.pie(df, values='tip', names='day')
 fig.show()
```

dados temporais, aqueles que passam ideia de cronologia. Uma solução natural é representarmos tais dados a partir do gráfico de linhas. Por exemplo:

```python
import plotly.express as px

 df = px.data.gapminder().query('continent=='Oceania'')
 fig = px.line(df, x='year', y='lifeExp', color='country')
 fig.show()
```
